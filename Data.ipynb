```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = [
   "data\Daily Shares  ETFs 2023 (2).csv",
   "data\Daily Shares  ETFs 2023 (4).csv",
   "data\Daily Shares  ETFs 2023 (5).csv",
   "data\Daily Shares  ETFs 2023 (6).csv",
   "data\Daily Shares  ETFs 2023 (7).csv",
   "data\Daily Shares  ETFs 2023 (8).csv",
   "data\Daily Shares  ETFs 2023 (9).csv",
   "data\Daily Shares  ETFs 2023 (10).csv",
]
```

```python
combined_data = []

for i, filepath in enumerate(files):
    df = pd.read_csv(filepath)
    df['Daily Date'] = pd.to_datetime(df['Daily Date'], dayfirst=True)
    cols = ['Closing Bid Price (GH¢)', 'Closing Offer Price (GH¢)']
    df[cols] = df[cols].fillna(0)
    
    combined_data.append(df)

combined_data =pd.concat(combined_data,ignore_index=True)
print(combined_data.head())    
```
